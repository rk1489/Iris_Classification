### Finding the optimum number of clusters and k-means clustering

```python
df.head()
```

```python
x = df.iloc[:, 0:-1].values
```

```python
from sklearn.cluster import KMeans
```

```python
# Checking for optimum wcss (within cluster sum of squares) i.e. reduction in variance
wcss = []

for i in range(1, 16):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
    kmeans.fit(x)
    wcss.append(kmeans.inertia_)
```

```python
# Plotting a line graph of the results to observe 'The elbow
plt.plot(range(1, 16), wcss, marker ="o", linewidth=2)
plt.title('The Elbow Method')
plt.xticks(np.arange(1,16,1))
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# Creating the KMeans classifier
kmeans = KMeans(n_clusters = 3, init = 'k-means++', max_iter = 300, n_init = 10, random_state = 0)
y_kmeans = kmeans.fit_predict(x)
```

```python
#Visualising the clusters
mpl.rcParams['figure.dpi'] = 250

plt.scatter(x[y_kmeans == 0, 0], x[y_kmeans == 0, 1], s = 30, c = 'red', label = 'Iris-setosa')
plt.scatter(x[y_kmeans == 1, 0], x[y_kmeans == 1, 1], s = 30, c = 'blue', label = 'Iris-versicolour')
plt.scatter(x[y_kmeans == 2, 0], x[y_kmeans == 2, 1], s = 30, c = 'green', label = 'Iris-virginica')

#Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], s = 80, marker='*', c = 'black', label = 'Centroids')

plt.legend(bbox_to_anchor=(1.08, 1.06),borderaxespad=0)
```

# Predicting species for raw data

```python
df.head()
```

```python
# Spliting data into training and test datasets
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, stratify = df['Species'], test_size=0.2, random_state=1) 
```

```python
X_train = train[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y_train = train.Species
X_test = test[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']]
y_test = test.Species
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

### Using KNN (K - Nearest Neighbour) algorithm

```python
# Applying KNN Algorithm
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors = 7, p = 2, metric='minkowski')
knn.fit(X_train, y_train)
```

```python
# Accuracy of the KNN classifier on training data
knn.score(X_train, y_train)
```

```python
# Accuracy of the KNN classifier on test data
knn.score(X_test, y_test)
```

```python
knn.predict(X_test)[:10]
```

```python
y_test.values[:10]
```

### Using SVC (Support Vector Classification) algorithm

```python
#Applying SVC (Support Vector Classification)
from sklearn.svm import SVC

svm = SVC(kernel='rbf', random_state=0, gamma=.10, C=1.0)
svm.fit(X_train, y_train)
```

```python
# Accuracy of the SVC classifier on training data
svm.score(X_train, y_train)
```

```python
# Accuracy of the SVC classifier on test data
svm.score(X_test, y_test)
```

```python
svm.predict(X_test)[:10]
```

```python
y_test.values[:10]
```

### Using Decision Tree algorithm

```python
# Applying Decision Tree
from sklearn import tree

tree = tree.DecisionTreeClassifier(criterion='gini')
tree.fit(X_train, y_train)
```

```python
# Accuracy of the Decision Tree classifier on training data
tree.score(X_train, y_train)
```

```python
# Accuracy of the Decision Tree classifier on test data
tree.score(X_test, y_test)
```

```python
tree.predict(X_test)[:10]
```

```python
y_test.values[:10]
```

### Using Random Forest algorithm

```python
# Using Random Forest
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# Accuracy of the Random Forest classifier on training data
rf.score(X_train, y_train)
```

```python
# Accuracy of the Random Forest classifier on test data
rf.score(X_test, y_test)
```

```python
rf.predict(X_test)[:10]
```
